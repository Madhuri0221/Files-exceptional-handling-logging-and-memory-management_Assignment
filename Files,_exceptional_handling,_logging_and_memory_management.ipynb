```python
# 2) What is exception handling in Python?

"""
Exception handling: While executing the code we might get few errors
those errors are to handled inorder to not terminate the execution
handling those errors is called exception handling

This can be done by using try-except blocks

The suspicious code ie., the code which is expected to throw error is
written in the try block
The error handling is done in the exception block

if the error is encountered in the try block then the code enters into the
except block and executes whats present in that block

if error is not encountered in try block then the flow doesnt go to the
except block, instead goes into the else block

we also have finally. which is always executed irrespective of either
exceptions occured or not

"""
try:
  print(4/0)
except Exception as e:
  print(f"Error occured: {e}")
```

```python
# 3) What is the purpose of the finally block in exception handling?

"""
To handle errors and not to stop execution flow because of the exceptions
occured

"""
try:
  print(4/0)
except Exception as e:
  print(f"Error occured: {e}")
else:
  print("Entered into else block,ie., error is not encountered")
finally:
  print("This finally block, always executed")
```

```python
# 7) How can you handle multiple exceptions in Python?

"""
By using single exception in single block - In 1 pair of try-except block
we can give single specific exception to be handled

By using multiple excpetions in single block - In 1 pair of try-except block
we can give multiple specific exceptions, but once one of the errors is
caught the other errors are not handles, execution exits the try-except block
the other error are never triggered

Handling multiple errors: We can use n number of try-except blocks to handle n
exceptions independently

"""

print("1 exception - 1 block")
try:
  print(5/0)
except ZeroDivisionError as e:
  print(f"Error occured: {e}")
print("")

print("2 exceptions - 1 block")
try:
  open("example123.txt", "r")
  print(5/0)
except (FileNotFoundError, ZeroDivisionError) as e:
  print(f"Error occured: {e}")
print("")

print("2 exceptions - 2 blocks")
try:
  print(5/0)
except ZeroDivisionError as e:
  print(f"Error occured: {e}")

try:
  open("example123.txt", "r")
except FileNotFoundError as e:
  print(f"Error occured: {e}")
```

```python
# 18) How would you implement a basic logging setup that logs to a file with rotation after 1MB

import logging
from logging.handlers import RotatingFileHandler

# Configure logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)  # Set logging level

# Create a rotating file handler
handler = RotatingFileHandler("app.log", maxBytes=1_000_000, backupCount=5)
handler.setFormatter(logging.Formatter("%(asctime)s - %(levelname)s - %(message)s"))

# Add handler to logger
logger.addHandler(handler)

# Generate some log messages
for i in range(10000):
    logger.info(f"Log entry {i}")
```

```python
# 21) Write a Python program that reads a file and prints the number of occurrences of a specific word?

def count_word_occurrences(file_path, word):
    try:
        with open(file_path, 'r') as file:
            content = file.read().lower()
            words = content.split()
            count = words.count(word.lower())
            print(f"The word '{word}' appears {count} times in the file.")
    except FileNotFoundError:
        print("Error: The file does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")


file_path = "D:\\Madhuri\\example.txt"
word_to_find = "water"
count_word_occurrences(file_path, word_to_find)

```

```python
# 23) Write a Python program that writes to a log file when an error occurs during file handling

import logging

logging.basicConfig(filename="error_log.txt",
                    level=logging.ERROR,
                    format="%(asctime)s - %(levelname)s - %(message)s")

def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            print("File content:\n", content)
    except FileNotFoundError:
        logging.error(f"File not found: {file_path}")
        print("Error: The file does not exist.")
    except Exception as e:
        logging.error(f"Unexpected error with file '{file_path}': {e}")
        print(f"An error occurred: {e}")

file_path = "D:\\Madhuri\\new2.txt"
read_file(file_path)
```
